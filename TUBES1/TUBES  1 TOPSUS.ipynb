```python
import matplotlib.pyplot as plt
import csv
from random import *
```

```python
with open('AALI.JK (1).csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
print(data[0])
```

```python
for i in range(100):
    naik = 0
    turun = 0
    returnNaik = 0
    returnTurun = 0
    for i in range(1,len(data)):
        close = float(data[i][4])
        if i < len(data) - 1:
            k = (float(data[i + 1][4]) - close)/ close
        if k > 0:
            naik += 1
            returnNaik += k * 1
            returnTurun += k * 0
        else:
            turun += 1
            returnNaik += k * 0
            returnTurun += k * 1
    closeAll = []
    simulasiAll = []
    dateAll = []
    for i in range(1,len(data)):
        dateAll.append(data[i][0])
        close = float(data[i][4])
        closeAll.append(close)
        if i == 1:
            simulasi = close
        else:
            if ran < p:
                simulasi = simulasi * (1 + u)
            else:
                simulasi = simulasi * (1 + d)
        simulasiAll.append(simulasi)
        if i < len(data) - 1:
            k = (float(data[i + 1][4]) - close)/ close
        p = naik / (naik + turun)
        satuminp = 1 - p
        u = returnNaik / naik
        d = returnTurun / turun
        ran = random()
    plt.plot(dateAll, simulasiAll, color='y')
plt.plot(dateAll, closeAll, color='blue')
plt.show()
```
